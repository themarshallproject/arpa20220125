```python
data_list = []
trouble = []

for pdf in pdfs:
    print(pdf)
    # creating a pdf file object 
    try: 
        pdfFileObj = open(pdf, "rb") 

        # creating a pdf reader object 
        pdfReader = PyPDF2.PdfFileReader(pdfFileObj) 

        # printing number of pages in pdf file 
        pages = pdfReader.numPages
        text=""
        for number in range(0, pages):    
            # creating a page object 
            pageObj = pdfReader.getPage(number) 

            # extracting text from page 
        #     print(pageObj.extractText()) 
            text = text + pageObj.extractText()
        # closing the pdf file object 

        data_list.append({
            "filename": pdf,
            "pages": pages,
            "text": text.replace("\n", "")
        })

        pdfFileObj.close() 
        
    except:
        trouble.append(pdf)
```

```python
df = pd.DataFrame(data_list)
```

## Natural Language Processing

```python
from nltk.tokenize import sent_tokenize
nltk.download('punkt')
```

```python
df = pd.read_csv("output_data/parsed_pdf_text.csv")
```

```python
df
```

```python
def tokenize_text(row):
    try:
        text = row["text"]
        tokenized_text = sent_tokenize(text)
    
        return tokenized_text
    except:
        print(row["filename"])
```

```python
df["tokenized_text"] = df.apply(tokenize_text, axis=1)
```

```python
from nltk.tokenize import word_tokenize
```

```python
def tokenize_word(row):
    try:
        text = row["text"]
        tokenized_word = word_tokenize(text)
    
        return tokenized_word
    except:
        print(row["filename"])
```

```python
df["tokenized_word"] = df.apply(tokenize_word, axis=1)
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
stop_words=set(stopwords.words("english"))
```

```python
def filter_words(row):
    try:
        tokenized_word = row["tokenized_word"]
        
        filtered_tokens=[]
        for w in tokenized_word:    
            if w not in stop_words:
                 filtered_tokens.append(w)
    
        return filtered_tokens
    except:
        print(row["filename"])
```

```python
df["filtered_tokens"] = df.apply(filter_words, axis=1)
```

```python
def find_similar_police(row):
    text = row["filtered_tokens"]
    
    try:
        if nltk.Text(text).concordance("shotspotter") == "no matches":
            pass
        else:
            print(nltk.Text(text).concordance("shotspotter"))
    except:
        pass
```

```python
df.apply(find_similar_police, axis=1)
```
