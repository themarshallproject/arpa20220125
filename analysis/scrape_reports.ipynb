```python
def download_interim_reports(row):
    url = row["Interim Report"]
    print(url)
    wait_for_element = 10
    
    driver.get(url)
    try:
        WebDriverWait(driver, wait_for_element).until(
            EC.element_to_be_clickable((By.CLASS_NAME, "downloadbutton"))
        )
        
        file_name = driver.find_elements_by_class_name("title")[0].text
        print(file_name)

        download_button = driver.find_element_by_class_name("downloadbutton")
        download_button.click()
        
    except TimeoutException as e:
        print("Wait Timed out")
        print(e)
    
    return file_name
```

```python
df_al = df[df["State/Territory"] == "Alabama"]
```

```python
df_al["interim_report_file_name"] = df_al.apply(download_interim_reports,\
                                               axis=1)
```
